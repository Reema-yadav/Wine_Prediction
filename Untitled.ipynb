```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('wine_fraud.csv')
df.head()
```

```python
df.info()
```

```python
df['quality'].unique()
```

```python
df['type'].unique()
```

```python
X = df.drop(columns = 'quality', axis =1)
```

```python
X 
```

```python
y = df['quality']
```

```python
y
```

```python
df[df['quality']=='Fraud']
```

```python
df.columns
```

```python
df.describe()
```

```python
X['type'].replace({'red':0, 'white':1}, inplace = True)
```

```python
X
```

```python
Y = y.replace({'Legit':0, 'Fraud':1})
```

```python
Y
```

```python
X.corr()
```

```python
standard = StandardScaler()
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size =0.2, random_state = 43) 
```

```python
x_train
```

```python
x_test
```

```python
standard = StandardScaler()
```

```python
x_train_scaled = standard.fit_transform(x_train)
x_test_scaled = standard.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter =1000, class_weight={0:0.3,1:0.7})
```

```python
model.fit(x_train_scaled, y_train)
```

```python
from sklearn.metrics import r2_score, f1_score, accuracy_score, confusion_matrix
```

```python
predicted = model.predict(x_test_scaled)
```

```python
accuracy = accuracy_score(y_test, predicted)
```

```python
accuracy
```

```python
confusion = confusion_matrix(y_test, predicted)
confusion
```
